# 모듈 불러오기

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
import numpy as np
print("TensorFlow version:", tf.__version__)
```

# 데이터 준비

```python
x_train = np.array( [20,21,22,23] ) # 온도
y_train = np.array( [40,42,44,46] ) # 판매량
```

# 모델 생성

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1))
```

# 모델 컴파일

```python
model.compile(loss='mean_squared_error',
                    optimizer='adam')
```

# 모델 학습

```python
model.fit(x=x_train, y=y_train, epochs=3000)
```

# 새로운 데이터 예측

```python
print( model.predict( np.array([ 20, 21, 22, 23 ]) ) )
```

# Weight 확인

```python
weights = model.get_weights()
가중치 = weights[0][0][0]
편향 = weights[1][0]
print('가중치(weight):', 가중치)
print('편향(bias):', 편향)
```

```python
result = 가중치 * x_train[0] + 편향
print(result)
```
