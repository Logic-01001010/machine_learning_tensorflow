# 모듈 불러오기

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
import numpy as np
```

# 데이터 준비

```python
x_train = np.array( [20,21,22,23] ) # 온도
y_train = np.array( [40,42,44,46] ) # 판매량
```

# 가설 세팅

```python
W = tf.Variable(tf.random.normal([1]), name="weight") # 가중치
b = tf.Variable(tf.random.normal([1]), name="bias") # 편향

hypothesis=x_train*W+b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

# 모델 생성

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1))
```

# 모델 컴파일

```python
model.compile(loss='mean_squared_error',
                    optimizer='adam')
```

# 모델 학습

```python
model.fit(x=x_train, y=y_train, epochs=10000)
```

# 새로운 데이터 예측

```python
print( model.predict( np.array([ 20, 21, 22, 23 ]) ) )
```
